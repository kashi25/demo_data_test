```python
import csv
import os
import numpy as np
import matplotlib.pyplot as plt

students = np.array([])

def add_student():
    student_number = input("Enter student number: ")
    surname = input("Enter surname: ")
    given_name = input("Enter given name: ")
    unit_mark = float(input("Enter unit mark: "))

    student = {
        'student_number': student_number,
        'surname': surname,
        'given_name': given_name,
        'unit_mark': unit_mark
    }

    students = np.append(students, student)
    print("Student added successfully.\n")

def search_students():
    search_key = input("Enter student number or name to search: ").lower()

    found_students = []

    for student in students:
        if (
            search_key in student['student_number'].lower()
            or search_key in student['surname'].lower()
            or search_key in student['given_name'].lower()
        ):
            found_students.append(student)

    if len(found_students) == 0:
        print("No matching students found.\n")
    else:
        print("Matching students:\n")
        for student in found_students:
            print_student_details(student)


def delete_student():
    search_key = input("Enter student number or name to delete: ").lower()

    deleted_students = []

    for student in students:
        if (
            search_key in student['student_number'].lower()
            or search_key in student['surname'].lower()
            or search_key in student['given_name'].lower()
        ):
            deleted_students.append(student)

    if len(deleted_students) == 0:
        print("No matching students found.\n")
    else:
        print("Deleted students:\n")
        for student in deleted_students:
            students = students[students != student]
            print_student_details(student)


def convert_to_grade(mark):
    if mark >= 80:
        return 'HD'
    elif mark >= 70:
        return 'D'
    elif mark >= 60:
        return 'C'
    elif mark >= 50:
        return 'P'
    else:
        return 'N'


def display_grade_students():
    grade = input("Enter unit grade (HD, D, C, P, or N): ")
    grade_students = []

    for student in students:
        student_grade = convert_to_grade(student['unit_mark'])
        if student_grade == grade:
            grade_students.append(student)

    if len(grade_students) == 0:
        print("No students with the given grade.\n")
    else:
        print("Students with grade", grade, ":\n")
        for student in grade_students:
            print_student_details(student)


def load_students_from_file(file_path):
    if not os.path.exists(file_path):
        print("File not found.\n")
        return

    with open(file_path, 'r') as file:
        reader = csv.DictReader(file)
        loaded_students = list(reader)

        for loaded_student in loaded_students:
            if not is_duplicate_student(loaded_student):
                students = np.append(students, loaded_student)

        print("Student records loaded successfully.\n")


def is_duplicate_student(student):
    for existing_student in students:
        if existing_student['student_number'] == student['student_number']:
            return True
    return False


def save_students_to_file(file_path):
    if os.path.exists(file_path):
        overwrite = input("File already exists. Do you want to overwrite it? (y/n): ")
        if overwrite.lower() != 'y':
            print("Operation canceled.\n")
            return

    with open(file_path, 'w', newline='') as file:
        fieldnames = ['student_number', 'surname', 'given_name', 'unit_mark']
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(students)

        print("Student records saved successfully.\n")


def display_grade_distribution():
    grade_counts = {}
    for student in students:
        grade = convert_to_grade(student['unit_mark'])
        grade_counts[grade] = grade_counts.get(grade, 0) + 1

    grades = list(grade_counts.keys())
    counts = list(grade_counts.values())

    plt.figure(figsize=(6, 6))
    plt.pie(counts, labels=grades, autopct='%1.1f%%')
    plt.title("Grade Distribution")
    plt.show()


def display_marks_distribution():
    marks_ranges = ['0-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80-89', '90-100']
    marks_counts = [0] * 8

    for student in students:
        mark = student['unit_mark']
        if mark >= 90:
            marks_counts[7] += 1
        elif mark >= 80:
            marks_counts[6] += 1
        elif mark >= 70:
            marks_counts[5] += 1
        elif mark >= 60:
            marks_counts[4] += 1
        elif mark >= 50:
            marks_counts[3] += 1
        elif mark >= 40:
            marks_counts[2] += 1
        elif mark >= 30:
            marks_counts[1] += 1
        else:
            marks_counts[0] += 1

    plt.figure(figsize=(10, 6))
    plt.bar(marks_ranges, marks_counts)
    plt.xlabel("Marks Range")
    plt.ylabel("Number of Students")
    plt.title("Marks Distribution")
    plt.show()


def display_menu():
    print("Menu:")
    print("1. Add a student")
    print("2. Search for students")
    print("3. Delete a student")
    print("4. Display students by grade")
    print("5. Load student records from a CSV file")
    print("6. Save student records to a CSV file")
    print("7. Display grade distribution")
    print("8. Display marks distribution")
    print("9. Quit")


def main():
    file_path = "students.csv"
    load_students_from_file(file_path)

    while True:
        display_menu()
        choice = input("Enter your choice (1-9): ")

        if choice == '1':
            add_student()
        elif choice == '2':
            search_students()
        elif choice == '3':
            delete_student()
        elif choice == '4':
            display_grade_students()
        elif choice == '5':
            file_path = input("Enter file path to load student records: ")
            load_students_from_file(file_path)
        elif choice == '6':
            file_path = input("Enter file path to save student records: ")
            save_students_to_file(file_path)
        elif choice == '7':
            display_grade_distribution()
        elif choice == '8':
            display_marks_distribution()
        elif choice == '9':
            print("Thank you for using the student management system.")
            break
        else:
            print("Invalid choice. Please try again.\n")


if __name__ == '__main__':
    main()
```
